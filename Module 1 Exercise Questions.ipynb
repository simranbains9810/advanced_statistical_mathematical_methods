# To calculate the mean:
# mean = (170+150+185+170+180) / 5

```python
mysum=0
mycount=0
for i in y:
  mysum+=i # Increate the sum by value (i)
  mycount+=1 # Increate the counter by 1

mean = mysum/mycount
print(mean)
```

"""**Question 3.1:** Let's calculate the variance and the standard deviation. Develop a Python script to calclualte the mean, variance and standard deviation without using any in-build functions or libraries. You can use only the **math.sqrt()** function.

<img src='https://drive.google.com/uc?id=1TILfsLPhdXLqyGIjILDLcNtvk-wACxyX' width="50" height="50"> 
> This is a challenging task, so take your time to study it.
"""

```python
import math

sumdif = 0
for height in y:
  # The following calculates the difference of each height from the mean
  # Then, we square this difference
  # Finally we sum all the square differences

  sumdif += (height-mean)**2 

# The variance is the sum of squared differences divided by the number of elements
variance=sumdif/mycount

# The standard deviation is the square root of the variance
std = math.sqrt(variance)

# Let's print everything
print("Mean:",mean)
print("Variance:",variance)
print("Standard deviation:",std)
```
